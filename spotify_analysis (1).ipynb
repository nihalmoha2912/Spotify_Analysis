 ## Importing Libraries and Loading Dataset

```python
num_df = df.select_dtypes(include=['int64', 'float64'])

Q1 = num_df.quantile(0.25)
Q3 = num_df.quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = (num_df < lower) | (num_df > upper)
print(len(outliers))

```

## Data Cleaning

```python
# dropping null and duplicate values
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python


numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    tenth_percentile = df[col].quantile(0.10)
    ninetieth_percentile = df[col].quantile(0.90)
    
    df[col] = np.where(df[col] < tenth_percentile, tenth_percentile,np.where(df[col] > ninetieth_percentile, ninetieth_percentile, df[col]))
```

```python
df.isnull().sum()
```

## Univariate Analysis

```python
for col in df.columns:
    
    if df[col].dtype in ['int','float']:
        plt.figure(figsize=(6,4))
        sns.displot(df[col], kde=True)
        plt.title(f"distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("frequency")
        plt.show()        
```

```python
for col in df.columns:
    
    if df[col].dtype in ['int','float']:
        plt.figure(figsize=(6,4))
        sns.boxplot(df[col])
        plt.title(f"distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("frequency")
        plt.show()
```

## Analysis

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
artist_stats1 = pd.DataFrame() 
for col in num_cols:
    artist_stats1[col] = df.groupby('track_artist')[col].mean()
result = pd.DataFrame()
for col in artist_stats1.columns:
    print(f"{col}")
    result = artist_stats1[col].sort_values(ascending=False)
    print(result.head())
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
album_name = pd.DataFrame()  # good start
for col in num_cols:
    album_name [col] = df.groupby('track_album_name')[col].mean()
result = pd.DataFrame()
for col in album_name.columns:
    print(f"{col}")
    result = album_name [col].sort_values(ascending=False)
    print(result.head())
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
playlist_genre1 = pd.DataFrame()  
for col in num_cols:
    playlist_genre1[col] = df.groupby('playlist_genre')[col].mean()
result = pd.DataFrame
for col in playlist_genre1.columns:
    print(f"{col}")
    result = playlist_genre1[col].sort_values(ascending=False)
    print(result.head())
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
playlist_subgenre1 = pd.DataFrame()  # good start
for col in num_cols:
    playlist_subgenre1[col] = df.groupby('playlist_subgenre')[col].mean().sort_values(ascending=False)
result = pd.DataFrame()
for col in playlist_subgenre1.columns:
    print(f"{col}")
    result = playlist_subgenre1[col].sort_values(ascending=False)
    print(result.head())
```

## Multivariate Analysis

```python
for col in playlist_subgenre1.columns:
    result = playlist_subgenre1[col].sort_values(ascending=False)
    plt.figure(figsize=(12,6))
    sns.barplot(x=result.index, y=result.values,palette="viridis")
    plt.title(f"Average {col} by Playlist Subgenre", fontsize=16, weight='bold')
    plt.xlabel("Playlist Subgenre", fontsize=12)
    plt.ylabel(f"Mean {col}", fontsize=12)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
for col in playlist_genre1.columns:
    result = playlist_genre1[col].sort_values(ascending=False)
    plt.figure(figsize=(12,6))
    sns.barplot(x=result.index, y=result.values,palette="viridis")
    plt.title(f"Average {col} by Playlist genre", fontsize=16, weight='bold')
    plt.xlabel("Playlist genre", fontsize=12)
    plt.ylabel(f"Mean {col}", fontsize=12)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
df_grouped = (df.groupby("track_artist")["track_popularity"].sum().sort_values(ascending=False).head(10).reset_index())
plt.figure(figsize=(12,6))
sns.barplot(df_grouped, x="track_artist", y="track_popularity",palette="viridis")
plt.title("top 10 artists by popularity")
plt.xlabel("Artist")
plt.ylabel("Popularity")
plt.tight_layout()
plt.show()
```

```python
df["track_album_release_date"] = pd.to_datetime(df["track_album_release_date"],errors="coerce")
df["year"]  = df["track_album_release_date"].dt.year
df["month"] = df["track_album_release_date"].dt.month
df["day"]   = df["track_album_release_date"].dt.day
df[['year','month','day']].head()
```

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
num_cols = [col for col in num_cols if col not in ["month", "day"]]
year1 = pd.DataFrame()
for col in num_cols:
    
    year1 = df.groupby("year")[num_cols].mean()

for col in year1.columns:
    print(f" {col}")
    result1 = year1[col].sort_values(ascending=False)
    print(result1.head())   
```

```python
for col in year1.columns:
    result1 = year1[col].sort_values(ascending=False)
    plt.figure(figsize=(12,6))
    sns.barplot(x=result1.index, y=result1.values,palette="viridis")
    plt.title(f"Average {col} by year", fontsize=16, weight='bold')
    plt.xlabel("Year", fontsize=12)
    plt.ylabel(f"Mean {col}", fontsize=12)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
